```python
st01Data = pd.read_csv("../dataset/st0001data.csv")
st02Data = pd.read_csv("../dataset/st0002data.csv")
```

```python
# x축은 시간 y축은 판매량으로 만들어보자 
```

```python
xValue = st01Data.WEEK #list로 갑싸는 것도 가능
yValue = st01Data.MEAN_VOLUME
```

```python
plt.plot(xValue,yValue)
```

```python
plt.xticks(range( min(xValue), max(xValue),3))
plt.plot(xValue,yValue)
```

```python
#차트 꾸미기 
plt.xticks(range( min(xValue), max(xValue),3))
plt.plot(xValue,yValue)
#x/y축 라벨 붙여주기 
plt.xlabel("week")
plt.ylabel("avg_volume")
# 차트제목
plt.title("weekly trend")
# 선색상, 종류, 굵기변경
plt.plot(xValue,xValue, "yo-",marker='*', lw=1.5)
```

```python
x2Value = st02Data.WEEK
y2Value = st02Data.MEAN_VOLUME
```

```python
plt.xlabel('WEEK')
plt.ylabel('AVG_VOLUME')
plt.plot(xValue,yValue,label = "ST0001") #1번째 차트 + 차트마다 라벨붙여주기 
plt.plot(x2Value,y2Value,"g--", label = "ST0002") #2번째 차트 
plt.legend(loc = 0)
```

```python
data = pd.read_csv("../dataset/feature_regression_example.csv")
```

```python
data
```

```python
data_2015 = data.loc [ data.YEAR == 2015]
```

```python
xData = data_2015.WEEK
yData = data_2015.QTY
```

```python
plt.plot(xData,yData,label="2015")
plt.legend(loc = 0)
plt.figure(figsize=(10 ,5)) #10은 x축 5는 y축 도화지 크기 
```

```python
fig = plt.figure(figsize=(10 ,5)) #10은 x축 5는 y축 도화지 크기 
ax1 = fig.add_subplot(211) #2행 1열의 1번째 영역에 그린다 
ax1.plot(xValue,yValue,"b--", label="ST0001")
ax1.legend(loc = 0)
ax2 = fig.add_subplot(212)
ax2.plot(x2Value,y2Value,"g" , label="ST0002")
ax2.legend(loc = 0)
```

```python
#데이터의 탐색적 분석 EDA

fig = plt.figure(figsize=(10 ,5)) #10은 x축 5는 y축 도화지 크기 
ax1 = fig.add_subplot(111) #2행 1열의 1번째 영역에 그린다 
ax1.plot(xValue,yValue,"b--", label="ST0001")
ax1.legend(loc = 1)
ax2 = ax1.twinx()
ax2.plot(xValue,y2Value,"g" , label="ST0002")
ax2.legend(loc = 2)
```

```python
doubleData = pd.read_csv("../dataset/feature_regression_example.csv")
```

```python
doubleData2015 = doubleData.loc [ data.YEAR == 2015]
```

```python
x = doubleData2015.WEEK
y = doubleData2015.QTY
y2 = doubleData2015.PRO_PERCENT
```

```python
#데이터의 탐색적 분석 EDA

fig = plt.figure(figsize=(10 ,5)) #10은 x축 5는 y축 도화지 크기 
ax1 = fig.add_subplot(111) #2행 1열의 1번째 영역에 그린다 
ax1.plot(x,y,"b--",marker='o')
ax2 = ax1.twinx()
ax2.plot(x,y2,"g" , label="2nd")
ax2.legend(loc = 2)

plt.ylabel("qty")
# 차트제목
plt.title("subPlot")
```
