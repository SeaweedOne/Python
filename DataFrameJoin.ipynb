JOIN 

```python
#라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
#파일 불러오기 
joinMaster = pd.read_csv("./dataset/join_region_master.csv", encoding="ms949")
joinSellout = pd.read_csv("./dataset/join_kopo_product_volume.csv", encoding="ms949")
```

```python
#testMaster변수에 joinMaster를 인덱스로 가지고옴. REGIONID가 A02 인 컬럼을 A01로 바꿈
testMaster = joinMaster.loc[0:1]
testMaster.loc[testMaster.REGIONID == "A02", "REGIONID"] = "A01"

testMaster
```

```python
#Inner Join (교집합 )
pd.merge(left = joinSellout,
        right= joinMaster,
        on="REGIONID",
        how="inner")
```

```python
#InnerJoin  
final = pd.merge(left=joinSellout,
        right=testMaster,
        on="REGIONID",
        how="inner")

final
```

```python
#Left Join ( 왼쪽 테이블의 모든 값 살림 )
pd.merge(left = joinMaster,
        right= joinSellout,
        on="REGIONID",
        how="left")
```

```python
#Right Join ( 왼쪽 테이블의 모든 값 살림 ) -> 잘 안씁니당
pd.merge(left = joinSellout,
        right= joinMaster,
        on="REGIONID",
        how="right")
```

```python
#데이터 병합 및 outer Join
```

```python
#joinMaster의 컬럼명 변경
#각 컬럼의 4번째줄에 새로운 데이터 삽입 
joinMaster = joinMaster.rename(columns={"REGIONID":"AP2ID"})
joinMaster.loc[4,"AP2ID"]="A05"
joinMaster.loc[4,"REGIONNAME"]="인도"
joinMaster
```

```python
#Outer Join ( 왼쪽/오른쪽 테이블의 모든 값 살림 ) 
pd.merge(left=joinSellout,
        right=joinMaster,
        left_on="REGIONID",
        right_on="AP2ID",
        how="outer")
```

데이터 만들기 

```python
columnList = ["REGIONID","PRODUCT","DATE","QTY"]

data1 = pd.DataFrame([["서울","AA카드",20210518,100],
 ["서울","AA카드",20210517,100],
 ["서울","AA카드",20210516,100],
 ["서울","AA카드",20210515,100],
 ["서울","AA카드",20210514,100]], columns=columnList)

#데이터를 카피해서 원하는 컬럼만 바꿔 넣어준다.
data2 = data1.copy()
data2.PRODUCT="BB카드"

data3 = data1.copy()
data3.REGIONID="부산"

data4 = data3.copy()
data4.PRODUCT = "BB카드"
```

```python
#만든 데이터들을 연속해서 붙이기
mergedData = data1.append(data2).append(data3).append(data4)
mergedData
```

실습

```python
promotionColumns = ["SUBSIDIARY","CARD_NAME","PROMOTION_NAME","PROMOTION_RATE"]

promotionData = pd.DataFrame([
    ["서울","AA카드","서울_프로모션",0.5],
    ["부산","BB카드","부산_프로모션",0.3],
    ["부산","CC카드","부산_프로모션",0.25]
],columns = promotionColumns)

promotionData
```

```python
# 조인의 키값은 여러개가 들어갈 수 있다.
# 지역으로만 키값이 들어가게되면 서울지역에 프로모션이 들어가지 않은 상품까지 프로모션이 표시된다.
# 항상 데이터를 살피며 조인하자!

pd.merge(left = promotionData,
        right= mergedData,
        left_on=("SUBSIDIARY","CARD_NAME"),
        right_on=("REGIONID","PRODUCT"),
        how="left")
```

```python
pd.merge(left = mergedData,
        right = promotionData,
        left_on = ("REGIONID","PRODUCT"),
        right_on = ("SUBSIDIARY","CARD_NAME"),
        how="left")
```
