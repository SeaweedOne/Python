```python
#라이브러리 불러오기
import pandas as pd
import numpy as np

#파일 불러오기 및 형 변환
selloutData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
selloutData = selloutData.astype({"REGIONID":str,
                                  "PRODUCT":str,
                                  "YEARWEEK":str,
                                  "QTY":float})

selloutData["YEAR"] , selloutData["WEEK"] = selloutData.YEARWEEK.str[:-2] , selloutData.YEARWEEK.str[-2:]

#분기별로 나눠 그룹 변수에 넣어줌
group1 = selloutData.loc[ selloutData.WEEK.astype(int) <=13 ]
group2 = selloutData.loc[ (selloutData.WEEK.astype(int)> 13) & (selloutData.WEEK.astype(int) <=26) ]
group3 = selloutData.loc[ (selloutData.WEEK.astype(int)> 26) & (selloutData.WEEK.astype(int) <=39) ]
group4 = selloutData.loc[ (selloutData.WEEK.astype(int)> 39) & (selloutData.WEEK.astype(int) <=52) ]

#groupKey를정의 
groupKey = ["REGIONID","YEAR","WEEK","PRODUCT"] 
#groupKey를 이용해 groupby 실행하고 MEAN컬럼에 평균값 넣어주기 / mean컬럼 -> qty_new로 재설정 
#1분기
groupedData1 = group1.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData1 = groupedData1.rename(columns = {"mean":"MEAN"})

#2분기
groupedData2 = group2.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData2 = groupedData2.rename(columns = {"mean":"MEAN"})

#3분기
groupedData3 = group3.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData3 = groupedData3.rename(columns = {"mean":"MEAN"})

#4분기
groupedData4 = group4.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData4 = groupedData4.rename(columns = {"mean":"MEAN"})

#결과출력 
print("1분기")
print(groupedData1)
print("\n2분기")
print(groupedData2)
print("\n3분기")
print(groupedData3)
print("\n4분기")
print(groupedData4)
```

```python
#라이브러리 불러오기
import pandas as pd
import numpy as np

#파일 불러오기 및 형 변환
selloutData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
selloutData = selloutData.astype({"REGIONID":str,
                                  "PRODUCT":str,
                                  "YEARWEEK":str,
                                  "QTY":float})

selloutData["YEAR"] , selloutData["WEEK"] = selloutData.YEARWEEK.str[:-2] , selloutData.YEARWEEK.str[-2:]

#분기별로 나눠 그룹 변수에 넣어줌
group1 = selloutData.loc[ selloutData.WEEK.astype(int) <=13 ]
group2 = selloutData.loc[ (selloutData.WEEK.astype(int)> 13) & (selloutData.WEEK.astype(int) <=26) ]
group3 = selloutData.loc[ (selloutData.WEEK.astype(int)> 26) & (selloutData.WEEK.astype(int) <=39) ]
group4 = selloutData.loc[ (selloutData.WEEK.astype(int)> 39) & (selloutData.WEEK.astype(int) <=52) ]

#분기 컬럼 추가해 분기별로 구분이 쉽게해줌
group1["QUARTER"] = 1
group2["QUARTER"] = 2
group3["QUARTER"] = 3
group4["QUARTER"] = 4

#groupKey를정의 
groupKey = ["REGIONID","PRODUCT" , "QUARTER"] 
#groupKey를 이용해 groupby 실행하고 MEAN컬럼에 평균값 넣어주기 / mean컬럼 -> qty_new로 재설정 
#1분기
groupedData1 = group1.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData1 = groupedData1.rename(columns = {"mean":"MEAN"})

#2분기
groupedData2 = group2.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData2 = groupedData2.rename(columns = {"mean":"MEAN"})

#3분기
groupedData3 = group3.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData3 = groupedData3.rename(columns = {"mean":"MEAN"})

#4분기
groupedData4 = group4.groupby(groupKey)["QTY"].agg(["mean"]).reset_index()
groupedData4 = groupedData4.rename(columns = {"mean":"MEAN"})

#결과출력 
print("분기별 평균 판매량")
print("\n1분기")
print(groupedData1)
print("\n2분기")
print(groupedData2)
print("\n3분기")
print(groupedData3)
print("\n4분기")
print(groupedData4)
```
